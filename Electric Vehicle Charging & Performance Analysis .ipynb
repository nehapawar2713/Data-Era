# Electric Vehicle Charging & Performance Analysis Using Python

```python
#importing libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#load the dataset
df=pd.read_csv("ELECTRIC_VEHICLE.csv")
df
```

```python
#unique values
df["Vehicle Model"].unique()
```

```python
#find the count of each category
df["Vehicle Model"].value_counts()
```

```python
#plot the chart
df['Vehicle Model'].value_counts().plot(kind='bar')
```

```python
#BOXPLOT FOR BATTERY CAPACITY
sns.boxplot(df['Battery Capacity (kWh)'])
```

```python
#how many BMWQ I3 Models are there in the dataset
bmw_i3_count=df[df['Vehicle Model'] =='BMW i3'].shape[0]
print(f'Number of BMW i3 models in the dataset: {bmw_i3_count}')
```

```python
# find the summary for the coloumn battery capacity.
df['Battery Capacity (kWh)'].describe()
```

```python
#mean of battery capacity
df['Battery Capacity (kWh)'].mean()
```

```python
#standard deviation of battery capacity
df['Battery Capacity (kWh)'].std()
```

```python
# how many vehciles are there whose battery capacity is more than the average battery capacity.
df[df["Battery Capacity (kWh)"]>df["Battery Capacity (kWh)"].mean()].count()
```

```python
# how many charging stations are there in san francisco
df[df['Charging Station Location']=='San Francisco'].count()
df
```

```python
#how many station 104 are in chicago.
df[(df['Charging Station Location']=='Chicago')&(df['Charging Station ID']=='Station_104')].count()
```

```python
# find the unique values of coloumn user type and find the count of each type , how many users are long distance traveller.
df[df['User Type']=='Long-Distance Traveller'].count()
```

```python
# Find how many casual drivers use DC fast charger
casual_dc_fast_charger_count=df[(df['User Type']=='Casual Driver')&(df['Charger Type']=='DC Fast Charger')].shape[0]
# Print the result
print(f"Number of Casual Drivers using DC Fast Charger: {casual_dc_fast_charger_count}")
```

```python
# write a code to make a scatter plot between battery capacity and temp.
sns.scatterplot(x=df['Battery Capacity (kWh)'],y=df['Temperature (°C)'],data=df)
```

```python
# find the correlation between battery capacity and temperature , what is the most common vehcile model in the dataset.
df['Battery Capacity (kWh)'].corr(df['Temperature (°C)'])
```

```python
# which model has the maximum battery capacity and what is the most common date of the week for charging stations.
max_battery_model = df.loc[df['Battery Capacity (kWh)'].idxmax(), 'Vehicle Model']
print(max_battery_model)
```
